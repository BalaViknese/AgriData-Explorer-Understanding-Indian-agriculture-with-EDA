```python
import mysql.connector

mydb = mysql.connector.connect(
  host="127.0.0.1",
  user="root",
  password="19Jul1999",
  database="indian_agri"  
)

mycursor = mydb.cursor()
```

```python
mycursor.execute("SHOW DATABASES")
for x in mycursor:
    print(x)
```

```python
mycursor.execute("""WITH top_states AS (
    SELECT state_name
    FROM agri_data
    GROUP BY state_name
    ORDER BY SUM(`rice_production_(1000_tons)`) DESC
    LIMIT 3
)
SELECT year, state_name, SUM(`rice_production_(1000_tons)`) AS total_rice_production
FROM agri_data
WHERE state_name IN (SELECT state_name FROM top_states)
GROUP BY year, state_name
ORDER BY year, total_rice_production DESC;""")
results = mycursor.fetchall()

for row in results:
    print(row)

mydb.commit()
```

```python
mycursor.execute("""SELECT dist_name AS district_name,
       MAX(CASE WHEN year = 2017 THEN `wheat_yield_(kg_per_ha)` END) AS yield_2017,
       MAX(CASE WHEN year = 2012 THEN `wheat_yield_(kg_per_ha)` END) AS yield_2012,
       (MAX(CASE WHEN year = 2017 THEN `wheat_yield_(kg_per_ha)` END) -
        MAX(CASE WHEN year = 2012 THEN `wheat_yield_(kg_per_ha)` END)) AS yield_increase
FROM agri_data
GROUP BY dist_name
HAVING yield_2017 IS NOT NULL AND yield_2012 IS NOT NULL
ORDER BY yield_increase DESC
LIMIT 5;""")
results = mycursor.fetchall()

for row in results:
    print(row)

mydb.commit()
```

```python
mycursor.execute("""SELECT state_name,
       MAX(CASE WHEN year = 2017 THEN `oilseeds_production_(1000_tons)` END) AS prod_2017,
       MAX(CASE WHEN year = 2012 THEN `oilseeds_production_(1000_tons)` END) AS prod_2012,
       ROUND(((MAX(CASE WHEN year = 2017 THEN `oilseeds_production_(1000_tons)` END) - 
               MAX(CASE WHEN year = 2012 THEN `oilseeds_production_(1000_tons)` END)) / 
               MAX(CASE WHEN year = 2012 THEN `oilseeds_production_(1000_tons)` END)) * 100, 2) AS growth_rate_pct
FROM agri_data
GROUP BY state_name
HAVING prod_2017 IS NOT NULL AND prod_2012 IS NOT NULL AND prod_2012 > 0
ORDER BY growth_rate_pct DESC
LIMIT 5;""")
results = mycursor.fetchall()

for row in results:
    print(row)
    
mydb.commit()
```

```python
mycursor.execute("""SELECT dist_name, `rice_area_(1000_ha)`, `rice_production_(1000_tons)`, `wheat_area_(1000_ha)`, `wheat_production_(1000_tons)`, `maize_area_(1000_ha)`, `maize_production_(1000_tons)`
FROM agri_data
WHERE year BETWEEN 2010 AND 2020;""")
results = mycursor.fetchall()

for row in results:
    print(row)
mydb.commit()
```

```python
mycursor.execute("""WITH top_cotton_states AS (
    SELECT state_name
    FROM agri_data
    GROUP BY state_name
    ORDER BY SUM(`cotton_production_(1000_tons)`) DESC
    LIMIT 5
)
SELECT year, state_name, SUM(`cotton_production_(1000_tons)`) AS total_cotton_production
FROM agri_data
WHERE state_name IN (SELECT state_name FROM top_cotton_states)
GROUP BY year, state_name
ORDER BY year, state_name;""")
results = mycursor.fetchall()

for row in results:
    print(row)
mydb.commit()
```

```python
mycursor.execute("""
    SELECT 
    dist_name AS district,
    state_name,
    predicted_groundnut_production_2020
FROM 
    agri_data
WHERE 
    predicted_groundnut_production_2020 IS NOT NULL
GROUP BY 
    dist_name, state_name, predicted_groundnut_production_2020
ORDER BY 
    predicted_groundnut_production_2020 DESC
LIMIT 10;""")

results = mycursor.fetchall()

for row in results:
    print(row)
mydb.commit()
```

```python
mycursor.execute("""SELECT year, ROUND(AVG(`maize_yield_(kg_per_ha)`), 2) AS avg_maize_yield
FROM agri_data
GROUP BY year
ORDER BY year;""")
results = mycursor.fetchall()

for row in results:
    print(row)
mydb.commit()
```

```python
mycursor.execute("""SELECT state_name, SUM(`oilseeds_area_(1000_ha)`) AS total_oilseeds_area
FROM agri_data
GROUP BY state_name
ORDER BY total_oilseeds_area DESC;""")
results = mycursor.fetchall()

for row in results:
    print(row)
mydb.commit()
```

```python
mycursor.execute("""SELECT dist_name, state_name, MAX(`rice_yield_(kg_per_ha)`) AS max_rice_yield
FROM agri_data
GROUP BY dist_name, state_name
ORDER BY max_rice_yield DESC
LIMIT 10;""")
results = mycursor.fetchall()

for row in results:
    print(row)
mydb.commit()
```

```python
mycursor.execute("""WITH top_states AS (
    SELECT state_name
    FROM agri_data
    GROUP BY state_name
    ORDER BY SUM(`wheat_production_(1000_tons)` + `rice_production_(1000_tons)`) DESC
    LIMIT 5
)
SELECT year, state_name,
       SUM(`wheat_production_(1000_tons)`) AS total_wheat_production,
       SUM(`rice_production_(1000_tons)`) AS total_rice_production
FROM agri_data
WHERE state_name IN (SELECT state_name FROM top_states)
  AND year >= (SELECT MAX(year) FROM agri_data) - 10
GROUP BY year, state_name
ORDER BY year, state_name;""")
results = mycursor.fetchall()

for row in results:
    print(row)
mydb.commit()
```
